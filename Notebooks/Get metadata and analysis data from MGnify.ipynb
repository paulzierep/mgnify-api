# Put it together and download the analysis data

```python
from jsonapi_client import Session
import pandas as pd
import requests
import os

study_accessions = ["MGYS00005611", "MGYS00001499"]
data_type = "Taxonomic assignments SSU"

data_output_folder = 'data/SSU'
os.makedirs(output_folder, exist_ok=True)

for study_accession in study_accessions:

    print(study_accession)

    # get df of metadata and urls to data 
    with Session("https://www.ebi.ac.uk/metagenomics/api/v1") as mgnify:

        dfs = []
        for r in mgnify.iterate(f'studies/{study_accession}/downloads'):
            df = pd.json_normalize(r.json)
            df['url'] = str(r.links.self)
            dfs.append(df)

    # df of all data that can be downloaded for this study
    main_df = pd.concat(dfs)

    # get specific data table
    url = main_df.loc[main_df["attributes.description.label"] == data_type, "url"].iloc[0]
    response = requests.get(url)

    data_output_path = os.path.join(output_folder, f"{study_accession}.txt")
    with open(data_output_path, "w") as f:
        f.write(response.text)

    # get analysis metadata
    with Session("https://www.ebi.ac.uk/metagenomics/api/v1") as mgnify:
        analyses = map(lambda r: r.json, mgnify.iterate(f'studies/{study_accession}/analyses'))
        analyses = pd.json_normalize(analyses)

    meta_data_output_path = os.path.join(output_folder, f"{study_accession}.metadata.txt")
    analyses.to_csv(meta_data_output_path)

    # TODO add metadata of each sample (geolocation, biome)
    # Ideally also the ENA accession (phenotype information)
```

```python
# merge dfs to have identical rows

def create_identical_index(output_folder, mode = "outer"):
    """
    Converts csv files with different indexes to files with identical and sorted indexes, using 
    either inner or outer join merge based on the index
    """

    dfs = []
    columns_per_file = {}
    for file in os.listdir(output_folder):
        file_path = os.path.join(output_folder,file)
        df = pd.read_csv(file_path, sep="\t", index_col=0)
        # df.set_index(0, inplace=True)
        df.index.name = "features"
        columns_per_file[file] = list(df.columns)
        dfs.append(df)
        print(df.shape)

    if mode == "outer":
        #features that do not exist in one of the dfs get filled with zero
        main_df = pd.concat(dfs, join=mode, axis = 1)
        main_df.fillna(0, inplace=True)
        print(main_df.shape)

    elif mode == "inner":
        #only features that exist in both dfs are kept
        main_df = pd.concat(dfs, join=mode, axis = 1)
        print(main_df.shape)

    # print(main_df.index.is_unique)

    new_output = output_folder + '_mod'
    os.makedirs(new_output, exist_ok=True)

    #write back to file
    for file, columns in columns_per_file.items():
        df = main_df.loc[:,columns]
        # file_name, file_ending = file.split(".")[0], file.split(".")[1]
        out_file = os.path.join(new_output, file)
        df.to_csv(out_file)

create_identical_index(output_folder, mode="outer")
```
